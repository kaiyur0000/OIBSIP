#importing required libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#IMPORTING DATA SET

Now we will analyse Before covid dataset

---



```python
# importing our data set(i had stored in google drive)
data_before_covid = pd.read_csv("/content/drive/MyDrive/oasis/datasetsoasis/Unemployment in India.csv")
df1=data_before_covid
```

#**EDA**

```python
df1.head()
```

```python
df1.tail(20)
```

```python
# AS WE CAN SEE THE ROWS FROM 754 TO 767 IS HAVING NULL VALUES IN ALL COLOUMN . SO WE HAVE TO DELETE IT.
# WITH THE HELP OF DROP FUNCTION , WE WILL DELETE ALL THE ROWS AND SAVE THE REMAINING ONE TO df1

df1=df1.drop(range(754, 768))
```

```python
df1.tail()
```

```python
df1.shape
```

```python
df1.info()
```

```python
#These are all the unique Region/state in our dataset
print(df1.Region.unique())
# total number of states
print(df1.Region.nunique())
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
# AS THE NULL ROWS ARE VERY LESS, WE CAN DROP IT. IT DO NOT IMPACT OUR DATA.
df1=df1.dropna()
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

#This is a BAR-GRAPH which shows our Estimated Unemployment rate according to different states

```python
x=df1["Region"]
y=df1[' Estimated Unemployment Rate (%)']
plt1unemployed = plt.figure(figsize=(40, 15))

plt.title("Umemployment Rate ACC To States", fontweight='bold' ,fontsize=28)

plt.xlabel("Region",fontweight='bold',fontsize=22)
plt.ylabel("Estimated Unemployment rate(%)",fontweight='bold',fontsize=22)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=22)
plt.bar(x,y, color='b',align='center')
```

#**Regions with Unemployment Rate Greater Than The Mean**

```python
mean_unemployment_rate = df1[" Estimated Unemployment Rate (%)"].mean()
high_unemployment_regions = df1[df1[" Estimated Unemployment Rate (%)"] > mean_unemployment_rate]
high_unemployment_region_names = high_unemployment_regions["Region"].unique()
print(high_unemployment_region_names)
```

#**UNEMPLOYMENT ACCORDING TO TIME**

```python
fig_after_covid = plt.figure(figsize = (30, 15))
plt.bar(df1[' Date'], df1[' Estimated Unemployment Rate (%)'])
plt.title('Unemployment RATE according to TIME',fontweight='bold' ,fontsize=28)
plt.xlabel('DATE',fontweight='bold' ,fontsize=22)
plt.ylabel('Unemployment Rate',fontweight='bold' ,fontsize=22)
plt.show()
```

#This PLOT Shows Employed Numbers With Respect To States

```python
a=df1[' Estimated Employed']
b=df1['Region']
plt1employed = plt.figure(figsize=(40, 15))

plt.title("Employment Rate ACC To States", fontweight='bold' ,fontsize=28)
plt.xlabel("Region",fontweight='bold',fontsize=22)
plt.ylabel("Estimated Employment rate(in le7)",fontweight='bold',fontsize=22)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=22)

plt.bar(b, a,color="g")
```

#**Employment Rate Acc To Time**

```python
fig_after_covid = plt.figure(figsize = (15, 6))
plt.bar(df1[' Date'], df1[' Estimated Employed'])
plt.title('Employment Rate According to Time',fontweight='bold' ,fontsize=28)
plt.xlabel('DATE',fontweight='bold' ,fontsize=18)
plt.ylabel('Employment Rate',fontweight='bold' ,fontsize=18)
plt.show()
```

# This shows the correlation between different columns in our dataset

```python
cr=df1.corr()
sns.heatmap(cr,annot=True,cmap="rocket")
```

THE GRAPH SHOWS THAT THERE IS NO CORRELATION BETWEEN TWO DIFF COLUMN'S, AND IT MAKES SENSE !

#This PLOT Shows Labour Participation According to Regions

```python
a=df1["Region"]
b=df1[" Estimated Labour Participation Rate (%)"]

plt1participation = plt.figure(figsize=(40, 20))

plt.title("Labour participation ACC To States", fontweight='bold' ,fontsize=28)
plt.xlabel("Region",fontweight='bold',fontsize=22)
plt.ylabel("Estimated Labour Participation Rate (%)",fontweight='bold',fontsize=22)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=22)

plt.scatter(a,b)
plt.show()
```

#**Regions with High participation than mean**

```python
mean_participation_rate = df1[" Estimated Labour Participation Rate (%)"].mean()
high_participation_regions = df1[df1[" Estimated Labour Participation Rate (%)"] > mean_participation_rate]
high_participation_region_names = high_participation_regions["Region"].unique()
print(high_participation_region_names)
```
